```python
import pandas as pd

# Load the dataset
url = 'https://gist.github.com/rhema/3b4b729d98978b9020d85c9b9e7c9dd6/raw/e0831e96129b2a0af0c696efe132cb8503c1e631/NormalWaterFlyingPokemon.csv'
all_data = pd.read_csv(url)

# Count the occurrences of each type
type_counts = all_data['Type'].value_counts()

# Calculate the prior probability of each type
total_samples = len(all_data)
prior_probabilities = type_counts / total_samples

# Display the result
print("Prior Probability of P(y) for each type:")
print(prior_probabilities)
```

```python
# Calculate the conditional probabilities for each feature value given the Pokemon type
conditional_probabilities = {}

for feature in ['HP', 'Attack', 'Defense']:
    conditional_probabilities[feature] = {}
    unique_values = all_data[feature].unique()
    for value in unique_values:
        conditional_probabilities[feature][value] = {}
        for pokemon_type in ['Normal', 'Water', 'Flying']:
            count = len(all_data[(all_data[feature] == value) & (all_data['Type'] == pokemon_type)])
            total_count_type = len(all_data[all_data['Type'] == pokemon_type])
            conditional_probabilities[feature][value][pokemon_type] = count / total_count_type

# Display the result
print("Subset of Likelihood for each feature value given the Pokemon type:")
for feature in ['HP', 'Attack', 'Defense']:
    print(f"For {feature}:")
    for value in conditional_probabilities[feature]:
        print(f"p({feature}={value} |Type=Normal) = {conditional_probabilities[feature][value]['Normal']:.3f}")
        print(f"p({feature}={value} |Type=Water) = {conditional_probabilities[feature][value]['Water']:.3f}")
        print(f"p({feature}={value} |Type=Flying) = {conditional_probabilities[feature][value]['Flying']:.3f}")
```

```python
# Calculate the prior probabilities of each type
priors = prior_probabilities.to_dict()

# Calculate the total probability of the features
total_probability = 0
for pokemon_type in ['Normal', 'Water', 'Flying']:
    probability = 1
    for feature in ['HP', 'Attack', 'Defense']:
        value = 'Mid' if feature == 'HP' else 'High' if feature == 'Attack' else 'Mid'
        probability *= conditional_probabilities[feature][value][pokemon_type]
    total_probability += probability * priors[pokemon_type]

# Calculate the posterior probability for each type given the features of Mandibuzz
posterior_probabilities = {}
for pokemon_type in ['Normal', 'Water', 'Flying']:
    probability = 1
    for feature in ['HP', 'Attack', 'Defense']:
        value = 'Mid' if feature == 'HP' else 'High' if feature == 'Attack' else 'Mid'
        probability *= conditional_probabilities[feature][value][pokemon_type]
    posterior_probabilities[pokemon_type] = probability * priors[pokemon_type] / total_probability

# Display the result
print("Posterior Probability for each Type given the Features of Mandibuzz:")
for pokemon_type, probability in posterior_probabilities.items():
    print(f"p(Type={pokemon_type} | Features) = {probability:.6f}")
```

```python
# Features of Mandibuzz
mandibuzz_features = {'HP': 'Mid', 'Attack': 'High', 'Defense': 'Mid'}

# Calculate the posterior probability for each type given the features of Mandibuzz
posterior_probabilities = {}
total_posterior_probability = 0

for pokemon_type in ['Normal', 'Water', 'Flying']:
    # Calculate p(Features|Type) for each type
    p_features_given_type = 1
    for feature, value in mandibuzz_features.items():
        p_features_given_type *= conditional_probabilities[feature][value][pokemon_type]

    # Calculate the posterior probability p(Type|Features) using Bayes' theorem
    posterior_probabilities[pokemon_type] = p_features_given_type * prior_probabilities[pokemon_type]
    total_posterior_probability += posterior_probabilities[pokemon_type]

# Normalize the posterior probabilities
for pokemon_type in posterior_probabilities:
    posterior_probabilities[pokemon_type] /= total_posterior_probability

# Print the posterior probabilities
print("Posterior Probabilities:")
for pokemon_type, probability in posterior_probabilities.items():
    print(f"p(Type={pokemon_type} | Features) = {probability:.4f}")

# Determine the predicted answer (type with the highest probability)
predicted_type = max(posterior_probabilities, key=posterior_probabilities.get)
print("\nPredicted Answer:", predicted_type)
```
